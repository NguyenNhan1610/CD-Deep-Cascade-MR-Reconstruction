# Sampling Masks

This jupyter notebook illustrates the usage of functions that create sampling masks. These functions are implemented at *Modules/sampling.py*.

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import sys

# Add my modules to path
MY_UTILS_PATH = "../Modules/"
if not MY_UTILS_PATH in sys.path:
    sys.path.append(MY_UTILS_PATH)

# import my modules    
import sampling 
```

## Uniform Sampling

```python
pattern_shape = (218,170)
factor = 0.20
nlines = 12
radius = 9

p1 = sampling.uniform_pattern(pattern_shape,factor,center=True,dim="1D",radius_nlines=nlines,direction="column")
p2 = sampling.uniform_pattern(pattern_shape,factor,center=True,dim="1D",radius_nlines=nlines,direction="row")
p3 = sampling.uniform_pattern(pattern_shape,factor,center=False,dim="1D",direction ="column")
p4 = sampling.uniform_pattern(pattern_shape,factor,center=False,dim="1D",direction="row")
p5 = sampling.uniform_pattern(pattern_shape,factor,center=True,dim="2D",radius_nlines=radius)
p6 = sampling.uniform_pattern(pattern_shape,factor,center=False,dim="2D",radius_nlines=radius)

print("Sampling rate")
print("Uniform 1D, column, centre:",p1.sum()/p1.size)
print("Uniform 1D, row, centre:",p2.sum()/p2.size)
print("Uniform 1D, column, no centre:",p3.sum()/p3.size)
print("Uniform 1D, row, mo centre:",p4.sum()/p4.size)
print("Uniform 2D, centre:",p5.sum()/p5.size)
print("Uniform 2D, no centre:",p6.sum()/p6.size)

plt.figure(dpi = 300)
plt.subplot(231)
plt.imshow(p1,cmap = "gray")
plt.axis("off")
plt.title("Uniform 1D, col, centre",size=8)
plt.subplot(232)
plt.imshow(p2,cmap = "gray")
plt.axis("off")
plt.title("Uniform 1D, col, centre",size=8)
plt.subplot(233)
plt.imshow(p3,cmap = "gray")
plt.axis("off")
plt.title("Uniform 1D, row, no centre",size=8)
plt.subplot(234)
plt.imshow(p4,cmap = "gray")
plt.axis("off")
plt.title("Uniform 1D, col, no centre",size=8)
plt.subplot(235)
plt.imshow(p5,cmap = "gray")
plt.axis("off")
plt.title("Uniform 2D centre",size=8)
plt.subplot(236)
plt.imshow(p6,cmap = "gray")
plt.axis("off")
plt.title("Uniform 2D, no centre",size=8)
plt.show()
```

## Gaussian Sampling

```python
pattern_shape = (218,170)
factor = 0.20
nlines = 12
radius = 9

p1 = sampling.gaussian_pattern(pattern_shape,factor,dim="1D",direction="column")
p2 = sampling.gaussian_pattern(pattern_shape,factor,dim="1D",direction="row")
p3 = sampling.gaussian_pattern(pattern_shape,factor,dim="2D")

print("Sampling rate")
print("Gaussian 1D, column:",p1.sum()/p1.size)
print("Gaussian 1D, row:",p2.sum()/p2.size)
print("Gaussian 2D:",p3.sum()/p3.size)

plt.figure(dpi = 300)
plt.subplot(131)
plt.imshow(p1,cmap = "gray")
plt.axis("off")
plt.title("Gaussian 1D, col",size=8)
plt.subplot(132)
plt.imshow(p2,cmap = "gray")
plt.axis("off")
plt.title("Gaussian 1D, row",size=8)
plt.subplot(133)
plt.imshow(p3,cmap = "gray")
plt.axis("off")
plt.title("Gaussian 2D",size=8)
plt.show()
```

## Poisson Disc Sampling

```python
pattern_shape = (218,170)
radius = 15

p1 = sampling.poisson_disc_pattern(pattern_shape,center = False,radius = 5, k  = 10, r = 1.71)
p2 = sampling.poisson_disc_pattern(pattern_shape,center = True,radius = 5, k  = 10, r = 1.715)

print("Sampling rate")
print("Poisson Disc, no centre:",p1.sum()/p1.size)
print("Poisson Disc Centre:",p2.sum()/p2.size)

plt.figure()
plt.subplot(121)
plt.imshow(p1,cmap = "gray")
plt.axis("off")
plt.title("Poisson Disc, no centre",size=8)
plt.subplot(122)
plt.imshow(p2,cmap = "gray")
plt.axis("off")
plt.title("Poisson Disc, centre",size=8)
plt.show()
```
